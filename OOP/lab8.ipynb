```python
import pygame as pg


pg.init()
screen = pg.display.set_mode((640, 480))
COLOR_INACTIVE = pg.Color('lightskyblue3')
COLOR_ACTIVE = pg.Color('dodgerblue2')
FONT = pg.font.Font(None, 32)


class UnluckyException (Exception):
    pass



class InputBox:

    def __init__(self, x, y, w, h, text=''):
        self.rect = pg.Rect(x, y, w, h)
        self.color = COLOR_INACTIVE
        self.text = text
        self.txt_surface = FONT.render(text, True, self.color)
        self.active = False

    def handle_event(self, event):
        if event.type == pg.MOUSEBUTTONDOWN:
            if self.rect.collidepoint(event.pos):
                self.active = not self.active
            else:
                self.active = False
            self.color = COLOR_ACTIVE if self.active else COLOR_INACTIVE
        if event.type == pg.KEYDOWN:
            if self.active:
                if event.key == pg.K_RETURN:
                    print(self.text)
                    self.text = ''
                elif event.key == pg.K_BACKSPACE:
                    self.text = self.text[:-1]
                else:
                    self.text += event.unicode
                self.txt_surface = FONT.render(self.text, True, self.color)

    def update(self):
        width = max(200, self.txt_surface.get_width()+10)
        self.rect.w = width

    def draw(self, screen, flag1, flag2, flag3,flag4, input_boxes):
        box1, box2 = input_boxes
        number1, number2 = box1.text, box2.text
        color = (255,255,255)

        screen.blit(self.txt_surface, (self.rect.x+5, self.rect.y+5))
        pg.draw.rect(screen, self.color, self.rect, 2)
        pg.draw.rect(screen,(255,0,0),(420,180,150,60))
        screen.blit(FONT.render('Introduceti', True, self.color), (430, 190))
        screen.blit(FONT.render('doua numere', True, self.color), (425, 210))
        
        if flag1:
            screen.blit(FONT.render('Fill the box', True, color), (140, 140))
        elif flag3:
            screen.blit(FONT.render('nu este numar', True, color), (140, 140))
            
        if flag2:
            screen.blit(FONT.render('Fill the box', True, color), (140, 340))
        elif flag4:
            screen.blit(FONT.render('nu este numar', True, color), (140, 340))
       
        if not (flag1 or flag2 or flag3 or flag4):
            screen.blit(FONT.render(str(int(number1)/int(number2)), True, color), (475, 260))

def divide(input_boxes,screen):
    color = (255,255,255)
    box1, box2 = input_boxes
    number1, number2 = box1.text, box2.text
    
    flag1, flag2, flag3,flag4 = False, False, False, False

    if len(number1) == 0:
        flag1 = True
    elif not number1.isdigit():
        flag3 = True
        
    if len(number2) == 0:
        flag2 = True
    elif not number2.isdigit():
        flag4 = True

    if flag1 or flag2 or flag3 or flag4:
        return flag1,flag2,flag3,flag4
    
    
    if number2 == '13':
        raise UnluckyException ("Oh no, the unlucky number")
    
    return False,False,False,False
    


def main():
    clock = pg.time.Clock()
    input_box1 = InputBox(100, 100, 140, 32)
    input_box2 = InputBox(100, 300, 140, 32)
    input_boxes = [input_box1, input_box2]
    done = False
    flag1, flag2, flag3,flag4 = True, True, False, False

    while not done:
        

        for event in pg.event.get():
            if event.type == pg.QUIT:
                done = True
            for box in input_boxes:
                box.handle_event(event)
            if event.type == pg.MOUSEBUTTONDOWN:
                if pg.mouse.get_pos()[0] >= 420 and pg.mouse.get_pos()[1] >= 180:
                    if pg.mouse.get_pos()[0] <= 570 and pg.mouse.get_pos()[1] <= 240:
                        flag1, flag2, flag3,flag4 = divide(input_boxes, screen);
        for box in input_boxes:
            box.update()

        screen.fill((30, 30, 30))
        for box in input_boxes:
            box.draw(screen, flag1, flag2, flag3,flag4,input_boxes)

        pg.display.flip()
        clock.tick(30)


if __name__ == '__main__':
    main()
    pg.quit()
```
