```python
import pygame
import os
import random
from nltk.corpus import words

pygame.init()


WIDTH, HEIGHT = 800, 500
win = pygame.display.set_mode((WIDTH, HEIGHT))
WHITE = (255,255,255)
BLACK = (0,0,0)
DIRPATH = os.getcwd()
SQUARE_WIDTH = 35
GAP = 20
LETTER_FONT = pygame.font.SysFont('comicsans', 40)
WORD_FONT = pygame.font.SysFont('comicsans', 40)

images = []
for i in range(7):
    image = pygame.image.load(DIRPATH+"\hangman" + str(i) + ".png")
    images.append(image)

letters = []
startx = 60
starty = 360
for i in range(26):
    x1 = startx + (SQUARE_WIDTH+GAP)*(i%13)
    x2 = x1 + SQUARE_WIDTH
    y1 = starty + (i//13) * (GAP + SQUARE_WIDTH)
    y2 = y1 + SQUARE_WIDTH
    letters.append([x1, x2, y1, y2, chr(65 + i), True])

def draw():
    win.fill(WHITE)

    # draw title
    text = WORD_FONT.render("HANGMAN", 1, BLACK)
    win.blit(text, (WIDTH/2 - text.get_width()/2, 20))

    # draw word
    display_word = ""
    for letter in word:
        if letter in already_guessed:
            display_word += letter + " "
        else:
            display_word += "_ "
    text = WORD_FONT.render(display_word, 1, BLACK)
    win.blit(text, (360, 200))

    # draw buttons
    for letter in letters:
        x1, x2, y1, y2, ltr, visible = letter
        if visible:
            pygame.draw.rect(win, BLACK, (x1, y1, SQUARE_WIDTH,SQUARE_WIDTH ))  # .circle(win, BLACK, (x, y), RADIUS, 3)
            text = LETTER_FONT.render(ltr, 1, WHITE)
            win.blit(text, (x1 + SQUARE_WIDTH/2 - text.get_width()/2, y1 + SQUARE_WIDTH/2 - text.get_height()/2))

    win.blit(images[hangman_status], (150, 100))
    pygame.display.update()

def display_message(message):
    pygame.time.delay(1000)
    win.fill(WHITE)
    text = LETTER_FONT.render(message, 1, BLACK)
    win.blit(text, (WIDTH/2 - text.get_width()/2, HEIGHT/2 - text.get_height()/2))
    pygame.display.update()
    pygame.time.delay(3000)


hangman_status = 0
already_guessed = []
english_words = words.words()
words = [word.upper() for word in english_words if len(word)<=10]
word = random.choice(words)


win = pygame.display.set_mode((WIDTH, HEIGHT))
def main():
    global hangman_status

    run = True

    while run:

        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                run = False
            if pygame.mouse.get_pressed()[0]:
                m_x, m_y = pygame.mouse.get_pos()
                for letter in letters:
                    x1, x2, y1, y2, ltr, visible = letter
                    if visible:
                        if x1<= m_x <= x2 and y1<=m_y<=y2:
                            letter[5] = False
                            already_guessed.append(ltr)
                            if ltr not in word:
                                hangman_status += 1
        
        draw()
        won = True
        for letter in word:
            if letter not in already_guessed:
                won = False
                break
        
        if won:
            #to hide the main window
            display_message("You WON!")
            break

        if hangman_status == 6:
            display_message("You LOST! The word was "+word)
            break

main()
pygame.quit()
```
